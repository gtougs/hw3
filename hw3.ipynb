```python

#obtain dataset from subject who's rawid is kirby906a_ax.img
kirby906a_ax_df = kirb21_df.loc[(kirb21_df.rawid == 'kirby906a_ax.img')]

#view what that looks like
kirby906a_ax_df.head(50)
```

```python
#now let's calculate this subject's total volume of type I level I telencephalon (left annd right)

#left
telencephalon_L_frame = kirby906a_ax_df.loc[(kirby906a_ax_df.roi == 'Telencephalon_L') & (kirby906a_ax_df.type == 1) & (kirby906a_ax_df.level == 1)]

telencephalon_L_volume = telencephalon_L_frame.volume.iloc[0]

print('Total Volume of Type 1 Level 1 Telencephalon (Left): ')
print(telencephalon_L_volume)
print('\n')


#we can do the same for the right
telencephalon_R_frame = kirby906a_ax_df.loc[(kirby906a_ax_df.roi == 'Telencephalon_R') & (kirby906a_ax_df.type == 1) & (kirby906a_ax_df.level == 1)]

telencephalon_R_volume = telencephalon_R_frame.volume.iloc[0]

print('Total Volume of Type 1 Level 1 Telencephalon (Right): ')
print(telencephalon_R_volume)
print('\n')
```

```python
#now let's calculate ICV and TBV fractions

ICV = kirby906a_ax_df.icv[kirby906a_ax_df.rawid == 'kirby906a_ax.img']
ICV = ICV.iloc[0]

print('Total Intracranial Volume (ICV): ')
print(ICV)
print('\n')

ICV_fraction = (telencephalon_L_volume + telencephalon_R_volume) / ICV

print('Telencephalon Fraction of ICV: ')
print(ICV_fraction)
print('\n')

TBV = kirby906a_ax_df.tbv[kirby906a_ax_df.rawid == 'kirby906a_ax.img']
TBV = TBV.iloc[0]

print('Total Brain Volume (TBV): ')
print(TBV)
print('\n')

TBV_fraction = (telencephalon_L_volume + telencephalon_R_volume) / TBV

print('Telencephalon Fraction of TBV: ')
print(TBV_fraction)
print('\n')
```

```python
all_regions_df = kirby906a_ax_df.groupby(['type', 'level'])[['volume']].sum()

# rename 'volume' column → 'total_volume'
all_regions_df = all_regions_df.rename(columns={'volume': 'total_volume'})

all_regions_df.head(15)

```

```python
#bar plot for type 1 level 2

kirb21_906_copy = kirby906a_ax_df.copy()

kirb21_906_copy = kirb21_906_copy.loc[(kirb21_906_copy.type == 1) & (kirb21_906_copy.level == 2)]

sns.barplot(x='roi', y='volume', data=kirb21_906_copy, hue='roi')
plt.xticks(rotation=90)   # rotate 90° so they’re vertical
plt.title(label='Relative Volume of Each Brain Region at Level 2 for Subject 906')
plt.tight_layout()
plt.show()
```

```python
#get cropped data frame donut chart for type 1 level 3 rois
kirb21_906_copy = kirby906a_ax_df.copy()
kirb21_906_copy = kirb21_906_copy.loc[
    (kirb21_906_copy.type == 1) & (kirb21_906_copy.level == 3)
]
#preview
kirb21_906_copy.head(10)
```

```python
vol = kirb21_906_copy["volume"].astype(float).fillna(0.0)

thr = 0.02  # 2% threshold by share of total
total = vol.sum()
mask_big = vol >= (thr * total)   # compare to 2% of total directly

big   = kirb21_906_copy.loc[mask_big].sort_values("volume", ascending=False)
small = kirb21_906_copy.loc[~mask_big].sort_values("volume", ascending=False)

# Labels/sizes
labels = big["roi"].tolist()
sizes  = big["volume"].astype(float).tolist()

# Residual "Other" so wedges sum to 100%
other = total - np.sum(sizes)
if other > 0:
    labels.append(f"Other (<{int(thr*100)}%)")
    sizes.append(other)

# Colors
base_colors = list(plt.cm.tab20.colors)
colors = (base_colors[:len(sizes)-1] + ["lightgray"]) if other > 0 else base_colors[:len(sizes)]

# Plot donut
fig, ax = plt.subplots(figsize=(10, 10))

def autopct_fmt(p):
    # p is already a percentage provided by matplotlib
    return f"{p:.1f}%" if p >= thr*100 else ""

wedges, texts, autotexts = ax.pie(
    sizes,
    labels=labels,
    autopct=autopct_fmt,
    startangle=90,
    wedgeprops={'width': 0.4},
    pctdistance=0.78,
    textprops={'fontsize': 8},
    colors=colors
)

ax.set_title("%ICV by Brain Region (Regions < 2% Collapsed as 'Other')")
ax.axis("equal")

# List collapsed regions with their true % (computed on the fly from volumes)
if not small.empty:
    small_text = "\n".join(
        f"{r}: {v/total*100:.1f}%"
        for r, v in zip(small["roi"], small["volume"])
    )
    ax.text(1.25, 0.0, f"Other (<{int(thr*100)}%):\n{small_text}",
            transform=ax.transAxes, va="center",
            bbox=dict(boxstyle="round,pad=0.4", fc="white", ec="lightgray"))

plt.tight_layout()
plt.show()
```

```python
#import gene expression data directly from github
gene_expression_df = pd.read_csv('https://raw.githubusercontent.com/jhu-advdatasci/2018/refs/heads/master/data/GSE5859_exprs.csv')
#confirm data is there
gene_expression_df.head(10)
```

```python
gene_expression_copy = gene_expression_df.copy()

#get the mean of the values in each row, excluding the patient anonymized ID column Unnamed: 0
rowmean = gene_expression_copy.drop('Unnamed: 0', axis=1).mean(axis=1)

#set up the resulting dataframe copied from the original dataset
gene_expression_df_pt1 = gene_expression_copy.copy()

#drop the first columns from the rsesulting dataframe so it it aligns with the mean calculation
gene_expression_df_pt1.drop('Unnamed: 0', axis=1, inplace=True)
#subtract mean from the original value in each column of each corresponding row
gene_expression_df_pt1 = gene_expression_df_pt1.sub(rowmean, axis=0)

# Add the 'Unnamed: 0' column back to the beginning of the resulting DataFrame so we can associate with individual ID.
gene_expression_df_pt1.insert(0, 'Unnamed: 0', gene_expression_copy['Unnamed: 0'])

#check to make sure the operation was correctly performed.
gene_expression_df_pt1.head(10)
```

##Task 3
###Part2

```python
# Create a copy to work with
gene_expression_copy = gene_expression_df.copy()

# Drop the 'Unnamed: 0' column again for mean calculation
colmean = gene_expression_copy.drop('Unnamed: 0', axis=1).mean(axis=1)

gene_expression_df_pt2 = gene_expression_df_pt1.drop('Unnamed: 0', axis=1).sub(colmean, axis=0)

gene_expression_df_pt2.insert(0, 'Unnamed: 0', gene_expression_copy['Unnamed: 0'])

# Check to make sure the operation was correctly performed.
gene_expression_df_pt2.head(10)
```

##Task 3
###Part 3

```python
# Create a copy to work with
gene_expression_copy = gene_expression_df.copy()

# Drop the 'Unnamed: 0' column again for mean calculation
colstdv = gene_expression_copy.drop('Unnamed: 0', axis=1).std(axis=0)

gene_expression_df_pt3 = gene_expression_df_pt2.drop('Unnamed: 0', axis=1).divide(colstdv, axis=1)

gene_expression_df_pt3.insert(0, 'Unnamed: 0', gene_expression_copy['Unnamed: 0'])

# Check to make sure the operation was correctly performed. See normalized values below.
gene_expression_df_pt3.head(10)
```

```python
healthcare_df = pd.read_csv('https://raw.githubusercontent.com/jhu-advdatasci/2018/master/data/KFF/healthcare-spending.csv', skiprows = 2)

#healthcare_df.head(10)

#we can see using the head preview above that other lines than states were imported to the frame

#thus, we can modify the frame to only include the list of known states.
states = {
    "Alabama","Alaska","Arizona","Arkansas","California","Colorado","Connecticut","Delaware",
    "District of Columbia","Florida","Georgia","Hawaii","Idaho","Illinois","Indiana","Iowa",
    "Kansas","Kentucky","Louisiana","Maine","Maryland","Massachusetts","Michigan","Minnesota",
    "Mississippi","Missouri","Montana","Nebraska","Nevada","New Hampshire","New Jersey",
    "New Mexico","New York","North Carolina","North Dakota","Ohio","Oklahoma","Oregon",
    "Pennsylvania","Rhode Island","South Carolina","South Dakota","Tennessee","Texas","Utah",
    "Vermont","Virginia","Washington","West Virginia","Wisconsin","Wyoming"
}
#with this function used in this df reset
healthcare_df = healthcare_df[healthcare_df["Location"].isin(states)]

#let's also remove the excess text from the titles to make them clean year numbers only
healthcare_df.columns = healthcare_df.columns.str.split('__').str[0]

#let's make sure this all looks correct
healthcare_df.head(10)
```

```python
#now we can melt the dataframe into a more useful format where the individual years and spend can be categorized
#at the state level
healthcare_plot = healthcare_df.melt(id_vars='Location', var_name='Year', value_name='Spend (in Million USD)')
healthcare_plot['Year'] = healthcare_plot['Year'].astype(int)
# Drop any rows where all year values == 0

#healthcare_plot.head(10)

#use plotly to create an interactive plot since some of the numbers will be
fig = px.line(healthcare_plot, x='Year', y='Spend (in Million USD)', color='Location',
              title='Healthcare Spend by State by Year (From 1991-2014)')

#since the lines will be interactive, let's hide the legend since it will look messy and confusing
fig.update_layout(showlegend=False)
fig.show()
```

```python
healthcare_copy = healthcare_df.copy()

# compute row-wise mean across all numeric columns
hc_mean = healthcare_copy.select_dtypes(include='number').mean(axis=1, skipna=True)

# add mean column at the END of the DataFrame
healthcare_copy['Mean'] = hc_mean

# preview
healthcare_copy.head(10)

fig, ax = plt.subplots(figsize=(30, 12))

healthcare_sorted = healthcare_copy.sort_values('Mean', ascending=False)

ax.bar(healthcare_sorted['Location'], healthcare_sorted['Mean'], label='Location', color='blue')

ax.set_ylabel('Average Spend (in Million USD)')
ax.set_xlabel('State')
ax.set_title('Average Healthcare Spend by State (From 1991-2014)')
ax.legend(title='State')
plt.xticks(rotation=90)
plt.tight_layout()
plt.figure(figsize=(20, 50))

plt.show()


```
